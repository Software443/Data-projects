# Matplotlib

##### you have been provided with a dataset that list Ohio State's leading cause of death from the year 2012 using the two data points cause of death and percentile, draw a pie chart to visualize the dataset.

```python
# import the required dataset
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Data from Ohio state, leading causes of death in the state from year 2012
#cause of deaths
cause = 'Chronic Diseases','Unintentional injuries','Alzheimers','Infuenza and pnemonia','Sepsis','others'
# percentile
percentile = [62,5,4,2,1,26]
```

## set the piechart plot properties

```python
# set figure size
plt.figure(figsize=(10,10))
#explode the largest pie in the data set
explode = (0.05,0.03,0.04,0.35,0.03,0.04)
# set pie chart properties
plt.pie(percentile,labels=cause,explode=explode,autopct='%1.1f%%',startangle=70)
# set axis equal to draw pie as circle
plt.axis('equal')
# set title of the pie chart
plt.title('OhioState - 2012: Leading Causes of Death')
# show the plot (pie chart)
plt.show()
```
